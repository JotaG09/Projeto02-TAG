#### Código em Python para converter os dados do grafo em imagem para representação, usando o arquivo .dot, criado pelo código em C++.

```python
import os
import graphviz
from IPython.display import display

# Caminho do ficheiro gerado pelo C++
dot_path = "docs/grafo_limpo_relatorio.dot"

# Verifica se o ficheiro existe antes de tentar renderizar
if os.path.exists(dot_path):
    with open(dot_path, "r") as f:
        dot_content = f.read()

    # Renderiza a interface visual do Grafo Bipartido na ecrã
    grafo = graphviz.Source(dot_content)
    display(grafo)

    # Exporta uma imagem PNG limpa para anexar ao relatório
    grafo.render("docs/Grafo_Alocacao_Final", format="png", cleanup=True)
    print("\nVisualização gerada com sucesso! Imagem guardada em: docs/Grafo_Alocacao_Final.png")
else:
    print(f"Erro: O ficheiro '{dot_path}' não foi encontrado.")
    print("Certifique-se de ter corrido a célula com '!make run' antes de renderizar o grafo.")
```
